# 03 — Shuffle

**Topics covered:**

1. **What is a shuffle, when does it happen, and why is it expensive** — serialization, disk I/O, network transfer, deserialization. Operations that trigger it: `groupBy`, `join`, `repartition`, `distinct`, `sort`.
2. **Minimizing and optimizing shuffle** — pre-partition on join keys, broadcast joins, `reduceByKey` over `groupByKey`, filter early, tune `spark.sql.shuffle.partitions`, bucketing.

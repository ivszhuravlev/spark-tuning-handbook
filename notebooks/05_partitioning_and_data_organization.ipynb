# 05 — Partitioning & Data Organization

**Topics covered:**

1. **`repartition()` vs. `coalesce()`** — `repartition()` = full shuffle for N even partitions; `coalesce()` = merge without full shuffle (can only reduce).
2. **Partitioning vs. bucketing** — partitioning = directory-based (partition pruning at read); bucketing = hash-based into fixed buckets (shuffle-free joins when both tables bucketed on the same key).

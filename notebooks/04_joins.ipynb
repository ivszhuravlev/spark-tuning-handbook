# 04 — Joins

**Topics covered:**

1. **Join strategies: Broadcast Hash, Sort-Merge, Shuffle Hash** — BHJ (small table broadcast, 10 MB threshold), SMJ (both sides shuffled + sorted, default), SHJ (shuffled + hashed, no sort). AQE dynamic switching.
2. **Broadcast joins in depth** — `autoBroadcastJoinThreshold`, `broadcast()` hint, trade-off between driver memory pressure and eliminating shuffle.
3. **Handling data skew in joins** — salting, AQE skew join optimization, splitting hot keys, broadcast when possible.
